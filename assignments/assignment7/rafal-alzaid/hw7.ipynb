```python
# distribution of final grades for math
# additionally, i am analyzing G3 grades because the data set outlines that they are the cumulative final grade combing G1 and G2
plt.figure(figsize=(8, 4))
sns.histplot(math_data['G3'], kde=True, bins=20)
plt.title('Distribution of Final Math Grades')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
# distribution of final grades for port
plt.figure(figsize=(8, 4))
sns.histplot(port_data['G3'], kde=True, bins=20)
plt.title('Distribution of Final Portuguese Grades')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
math_data['pass'] = math_data['G3'].apply(lambda x: 'pass' if x >= 10 else 'fail')
plt.figure(figsize=(8, 4))
sns.countplot(x='pass', data=math_data)
plt.title('Count of Students Passing vs. Failing (MATH)')
plt.xlabel('Pass/Fail')
plt.ylabel('Number of Students')
plt.show()
```

#3. Hypothesis Formulation:


###**Hypothesis 1:** There is a correlation between final grades and the time the student has studied.

```python
combined_data = pd.concat([math_data, port_data])

plt.figure(figsize=(12, 6))
sns.boxplot(x='studytime', y='G3', hue='Subject', data=combined_data)
plt.title('Final Grades by Study Time: Math vs Portuguese')
plt.xlabel('Study Time (1= <2 hours, 2=2 to 5 hours, 3=5 to 10 hours, 4= >10 hours)')
plt.ylabel('Final Grade (G3)')
plt.legend(title='Subject')
plt.show()
```

As the box plot shows, there is a gradual increase in final grades as the student studies for more time.

###**Hypothesis 2:** There is a correlation between final grades and which school has better reputation/performance.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='school', y='G3', hue='Subject', data=combined_data)
plt.title('Comparison of Final Grades by School and Subject')
plt.xlabel('School')
plt.ylabel('Final Grade (G3)')
plt.legend(title='Subject')
plt.grid(True)
plt.show()
```

It could be concluded that the reputation or performance in terms of final grades might be slightly better for GP in Mathematics and for MS in Portuguese.
However, the differences are not dramatic, indicating that both schools perform relatively similarly across the two subjects.
